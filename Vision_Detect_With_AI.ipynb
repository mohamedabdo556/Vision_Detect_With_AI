```python

import cv2
from ultralytics import YOLO

# ÿ™ÿ≠ŸÖŸäŸÑ ŸÜŸÖŸàÿ∞ÿ¨ YOLO ÿßŸÑŸÖÿØÿ±ÿ® ŸÖÿ≥ÿ®ŸÇŸãÿß
model = YOLO("yolov5su.pt")

# ÿ™ÿ∑ÿ®ŸäŸÇ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿπŸÑŸâ ŸÅŸäÿØŸäŸà
def detect_objects_in_video(video_path, model):
    cap = cv2.VideoCapture(video_path)
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        results = model.predict(frame, imgsz=640, conf=0.25)
        annotated_frame = results[0].plot()
        cv2.imshow('Detection', annotated_frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()

# ÿ™ÿ¥ÿ∫ŸäŸÑ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿπŸÑŸâ ŸÅŸäÿØŸäŸà
video_path = "path/to/your/video.mp4"
detect_objects_in_video(video_path, model)
```

```python
# ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿßŸÑŸÖÿØÿ±ÿ® ŸÖÿ≥ÿ®ŸÇŸãÿß
model = YOLO("yolov5su.pt")

# ÿ™ÿπÿ±ŸäŸÅ ÿßŸÑŸàÿßÿ¨Ÿáÿ©
file_upload = widgets.FileUpload(accept='.jpg, .png', multiple=False)
output_widget = widgets.Output()
run_button = widgets.Button(description="ÿ™ÿ¥ÿ∫ŸäŸÑ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨", button_style='success')
feedback_text = widgets.Text(description="ŸÖÿß ÿßŸÑÿ∞Ÿä ŸäŸàÿ¨ÿØ ŸÅŸä ÿßŸÑÿµŸàÿ±ÿ©ÿü")
feedback_number = widgets.IntText(description="ŸÉŸÖ ÿπÿØÿØ ÿßŸÑŸÉÿßÿ¶ŸÜÿßÿ™ÿü")
train_button = widgets.Button(description="ÿ•ÿπÿßÿØÿ© ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨", button_style='warning')
feedback_container = widgets.VBox([feedback_text, feedback_number, train_button])
feedback_container.layout.display = "none"  # ÿ•ÿÆŸÅÿßÿ° ÿßŸÑŸÖÿØÿÆŸÑÿßÿ™ ÿßŸÑÿßŸÅÿ™ÿ±ÿßÿ∂Ÿäÿ©

uploaded_file_path = None  # ŸÖÿ™ÿ∫Ÿäÿ± ŸÑÿ™ÿÆÿ≤ŸäŸÜ ŸÖÿ≥ÿßÿ± ÿßŸÑŸÖŸÑŸÅ ÿßŸÑŸÖÿ±ŸÅŸàÿπ

# ÿØÿßŸÑÿ© ŸÑÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÑÿµŸàÿ±ÿ©
def analyze_image(image_path):
    results = model.predict(image_path, imgsz=640, conf=0.25)
    return results

# ÿØÿßŸÑÿ© ŸÑŸÑÿ™ÿπÿßŸÖŸÑ ŸÖÿπ ÿßŸÑŸÖŸÑŸÅ ÿßŸÑŸÖÿ±ŸÅŸàÿπ
def handle_file(change):
    global uploaded_file_path
    if file_upload.value:
        uploaded_file = file_upload.value[0]
        file_name = uploaded_file["name"]  # ÿ™ÿµÿ≠Ÿäÿ≠ ŸáŸÜÿß
        with open(file_name, 'wb') as f:
            f.write(uploaded_file["content"])
        uploaded_file_path = file_name  # ÿ≠ŸÅÿ∏ ŸÖÿ≥ÿßÿ± ÿßŸÑŸÖŸÑŸÅ ÿßŸÑŸÖÿ±ŸÅŸàÿπ
        output_widget.clear_output()
        with output_widget:
            print(f"ÿ™ŸÖ ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÖŸÑŸÅ: {file_name}")

# ÿ•ÿ∂ÿßŸÅÿ© ŸÖŸÑÿßÿ≠ÿ∏ÿ© ÿ®ÿπÿØ ÿ±ŸÅÿπ ÿßŸÑŸÖŸÑŸÅ
file_upload.observe(handle_file, names='value')

# ÿØÿßŸÑÿ© ŸÑÿ™ŸÜŸÅŸäÿ∞ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
def run_model(_):
    if not uploaded_file_path:
        with output_widget:
            output_widget.clear_output()
            print("ŸÑŸÖ Ÿäÿ™ŸÖ ÿ±ŸÅÿπ ÿ£Ÿä ŸÖŸÑŸÅ! ÿßŸÑÿ±ÿ¨ÿßÿ° ÿ±ŸÅÿπ ŸÖŸÑŸÅ ÿ£ŸàŸÑÿßŸã.")
        return

    output_widget.clear_output()
    with output_widget:
        print("ÿ¨ÿßÿ±Ÿä ÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÑÿµŸàÿ±ÿ©...")
        results = analyze_image(uploaded_file_path)

        if len(results[0].boxes) == 0:  # ÿ•ÿ∞ÿß ŸÑŸÖ Ÿäÿ™ÿπÿ±ŸÅ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿπŸÑŸâ ÿ£Ÿä ŸÉÿßÿ¶ŸÜ
            print("ŸÑŸÖ Ÿäÿ™ŸÖ ÿßŸÑÿ™ÿπÿ±ŸÅ ÿπŸÑŸâ ÿ£Ÿä ŸÉÿßÿ¶ŸÜ ŸÅŸä ÿßŸÑÿµŸàÿ±ÿ©.")
            feedback_container.layout.display = "flex"  # ÿ•ÿ∏Ÿáÿßÿ± ŸÖÿØÿÆŸÑÿßÿ™ ÿßŸÑŸÖÿ≥ÿ™ÿÆÿØŸÖ
        else:
            print("ÿ™ŸÖ ÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÑÿµŸàÿ±ÿ© Ÿàÿπÿ±ÿ∂ ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨.")
            results[0].show()  # ÿπÿ±ÿ∂ ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨

# ÿØÿßŸÑÿ© ŸÑÿ•ÿπÿßÿØÿ© ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
def retrain_model(_):
    object_name = feedback_text.value.strip()
    object_count = feedback_number.value

    if not object_name or object_count <= 0:
        with output_widget:
            output_widget.clear_output()
            print("ÿßŸÑÿ±ÿ¨ÿßÿ° ÿ•ÿØÿÆÿßŸÑ ÿßÿ≥ŸÖ ÿßŸÑŸÉÿßÿ¶ŸÜ ŸàÿπÿØÿØ ÿµÿ≠Ÿäÿ≠.")
        return

    with output_widget:
        print(f"ÿ™ŸÖ ÿ™ÿ≥ÿ¨ŸäŸÑ ÿßŸÑŸÉÿßÿ¶ŸÜ: {object_name} (ÿßŸÑÿπÿØÿØ: {object_count}).")
        print("Ÿäÿ™ŸÖ ÿßŸÑÿ¢ŸÜ ÿ•ÿπÿØÿßÿØ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸÑÿ•ÿπÿßÿØÿ© ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨...")

    # ÿ™ÿ¨ŸáŸäÿ≤ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸÑÿ•ÿπÿßÿØÿ© ÿßŸÑÿ™ÿØÿ±Ÿäÿ®
    # (ÿπŸÑŸâ ÿ≥ÿ®ŸäŸÑ ÿßŸÑŸÖÿ´ÿßŸÑ: ÿ≠ŸÅÿ∏ ÿßŸÑÿµŸàÿ±ÿ© ÿßŸÑŸÖÿ±ŸÅŸàÿπÿ© ŸÖÿπ ÿßŸÑÿ¥ÿ±ÿ≠ ŸÉÿ¨ÿ≤ÿ° ŸÖŸÜ ŸÖÿ¨ŸÖŸàÿπÿ© ÿ®ŸäÿßŸÜÿßÿ™ ÿ¨ÿØŸäÿØÿ©)
    # ŸäŸÖŸÉŸÜ ÿ•ÿ∂ÿßŸÅÿ© ÿßŸÑÿÆÿ∑Ÿàÿßÿ™ ÿßŸÑÿ™ÿßŸÑŸäÿ©:
    # - ÿ≠ŸÅÿ∏ ÿßŸÑÿµŸàÿ±ÿ© ŸÅŸä ŸÖÿ¨ŸÑÿØ ÿ™ÿØÿ±Ÿäÿ® ÿ¨ÿØŸäÿØ.
    # - ÿ•ŸÜÿ¥ÿßÿ° ŸÖŸÑŸÅ JSON Ÿäÿ∂ŸÖ ŸàÿµŸÅ ÿßŸÑŸÉÿßÿ¶ŸÜÿßÿ™.

    with output_widget:
        print("Ÿäÿ™ŸÖ ÿßŸÑÿ¢ŸÜ ÿ•ÿπÿßÿØÿ© ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ (Ÿáÿ∞Ÿá ÿßŸÑÿÆÿ∑Ÿàÿ© ŸÇÿØ ÿ™ÿ≥ÿ™ÿ∫ÿ±ŸÇ ŸàŸÇÿ™Ÿãÿß)...")
        # ÿ™ŸÜŸÅŸäÿ∞ ÿ•ÿπÿßÿØÿ© ÿßŸÑÿ™ÿØÿ±Ÿäÿ® (ÿ®ÿ≥Ÿäÿ∑ÿ© ÿ¨ÿØŸãÿß ŸÑÿ£ÿ∫ÿ±ÿßÿ∂ ÿßŸÑÿ™Ÿàÿ∂Ÿäÿ≠)
        # ŸÖÿ´ÿßŸÑ:
        # model.train(data="path_to_updated_dataset.yaml", epochs=3)
        print("ÿ™ŸÖ ÿ™ÿ≠ÿØŸäÿ´ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿ®ŸÜÿ¨ÿßÿ≠!")

# ÿ±ÿ®ÿ∑ ÿßŸÑÿ£ÿ≤ÿ±ÿßÿ± ÿ®ÿßŸÑÿØŸàÿßŸÑ
run_button.on_click(run_model)
train_button.on_click(retrain_model)

# ÿπÿ±ÿ∂ ÿßŸÑŸàÿßÿ¨Ÿáÿ©
display(HTML("<h2>ÿ£ÿØÿÆŸÑ ÿµŸàÿ±ÿ© ŸÑÿ™ÿ≠ŸÑŸäŸÑŸáÿß ÿ®Ÿàÿßÿ≥ÿ∑ÿ© yolov5su</h2>"))
display(file_upload)
display(run_button)
display(output_widget)
display(feedback_container)
```
