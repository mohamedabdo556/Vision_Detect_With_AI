```python

import cv2
from ultralytics import YOLO

# تحميل نموذج YOLO المدرب مسبقًا
model = YOLO("yolov5su.pt")

# تطبيق النموذج على فيديو
def detect_objects_in_video(video_path, model):
    cap = cv2.VideoCapture(video_path)
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        results = model.predict(frame, imgsz=640, conf=0.25)
        annotated_frame = results[0].plot()
        cv2.imshow('Detection', annotated_frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()

# تشغيل النموذج على فيديو
video_path = "path/to/your/video.mp4"
detect_objects_in_video(video_path, model)
```

```python
# تحميل النموذج المدرب مسبقًا
model = YOLO("yolov5su.pt")

# تعريف الواجهة
file_upload = widgets.FileUpload(accept='.jpg, .png', multiple=False)
output_widget = widgets.Output()
run_button = widgets.Button(description="تشغيل النموذج", button_style='success')
feedback_text = widgets.Text(description="ما الذي يوجد في الصورة؟")
feedback_number = widgets.IntText(description="كم عدد الكائنات؟")
train_button = widgets.Button(description="إعادة تدريب النموذج", button_style='warning')
feedback_container = widgets.VBox([feedback_text, feedback_number, train_button])
feedback_container.layout.display = "none"  # إخفاء المدخلات الافتراضية

uploaded_file_path = None  # متغير لتخزين مسار الملف المرفوع

# دالة لتحليل الصورة
def analyze_image(image_path):
    results = model.predict(image_path, imgsz=640, conf=0.25)
    return results

# دالة للتعامل مع الملف المرفوع
def handle_file(change):
    global uploaded_file_path
    if file_upload.value:
        uploaded_file = file_upload.value[0]
        file_name = uploaded_file["name"]  # تصحيح هنا
        with open(file_name, 'wb') as f:
            f.write(uploaded_file["content"])
        uploaded_file_path = file_name  # حفظ مسار الملف المرفوع
        output_widget.clear_output()
        with output_widget:
            print(f"تم تحميل الملف: {file_name}")

# إضافة ملاحظة بعد رفع الملف
file_upload.observe(handle_file, names='value')

# دالة لتنفيذ النموذج
def run_model(_):
    if not uploaded_file_path:
        with output_widget:
            output_widget.clear_output()
            print("لم يتم رفع أي ملف! الرجاء رفع ملف أولاً.")
        return

    output_widget.clear_output()
    with output_widget:
        print("جاري تحليل الصورة...")
        results = analyze_image(uploaded_file_path)

        if len(results[0].boxes) == 0:  # إذا لم يتعرف النموذج على أي كائن
            print("لم يتم التعرف على أي كائن في الصورة.")
            feedback_container.layout.display = "flex"  # إظهار مدخلات المستخدم
        else:
            print("تم تحليل الصورة وعرض النتائج.")
            results[0].show()  # عرض النتائج

# دالة لإعادة تدريب النموذج
def retrain_model(_):
    object_name = feedback_text.value.strip()
    object_count = feedback_number.value

    if not object_name or object_count <= 0:
        with output_widget:
            output_widget.clear_output()
            print("الرجاء إدخال اسم الكائن وعدد صحيح.")
        return

    with output_widget:
        print(f"تم تسجيل الكائن: {object_name} (العدد: {object_count}).")
        print("يتم الآن إعداد البيانات لإعادة تدريب النموذج...")

    # تجهيز البيانات لإعادة التدريب
    # (على سبيل المثال: حفظ الصورة المرفوعة مع الشرح كجزء من مجموعة بيانات جديدة)
    # يمكن إضافة الخطوات التالية:
    # - حفظ الصورة في مجلد تدريب جديد.
    # - إنشاء ملف JSON يضم وصف الكائنات.

    with output_widget:
        print("يتم الآن إعادة تدريب النموذج (هذه الخطوة قد تستغرق وقتًا)...")
        # تنفيذ إعادة التدريب (بسيطة جدًا لأغراض التوضيح)
        # مثال:
        # model.train(data="path_to_updated_dataset.yaml", epochs=3)
        print("تم تحديث النموذج بنجاح!")

# ربط الأزرار بالدوال
run_button.on_click(run_model)
train_button.on_click(retrain_model)

# عرض الواجهة
display(HTML("<h2>أدخل صورة لتحليلها بواسطة yolov5su</h2>"))
display(file_upload)
display(run_button)
display(output_widget)
display(feedback_container)
```
